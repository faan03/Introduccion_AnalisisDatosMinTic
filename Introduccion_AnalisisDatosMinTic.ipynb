##Ordenamiento Datos

```python
# Luego de cargar datos, se ordenan por peso/numero del bloque
dfData= dfData.sort_values(by=['Height'])#,ascending=False) 
#dfData= dfData.sort_values(by=['Number of Transactions']) 
dfData
```

##Filtrado de Datos

```python
#Nombre de mineros

lMinersBlocks = dfData['Miner Name']

lMinersBlocks = dfData['Miner Name'].unique()
lMinersBlocks = list(lMinersBlocks)
print(lMinersBlocks)
print(len(lMinersBlocks))
```

```python
#Filtro por Ao
#dfData[dfData['Year']==2021]
dfData[dfData['Year']==2021]['Date']
dfData[dfData['Year']==2021][['Miner Name','Date']]
```

```python
#dfData['Miner Name']=="Poolin"
#dfData[dfData['Miner Name']=="Poolin"]
dfData[dfData['Miner Name'].isin(["Poolin",'F2Pool']) ]
#['Number of Transactions']
```

```python
#dfData.groupby(['Year','Miner Name'])['Height'].count()
dfData.groupby(['Year','Miner Name'])['Height'].count().rename('minedBlock').reset_index()
```

```python
gTransactions = dfData.groupby(['Year', 'Miner Name'])['Number of Transactions'].mean().rename('Avg Number of Transactions').reset_index()
print("gTransactions")
print(gTransactions)

gDifficulty = dfData.groupby(['Year', 'Miner Name'])['Difficulty'].mean().rename('Avg Difficulty').reset_index()
print("gDifficulty")
print(gDifficulty)

#Unir los DataFrame
newDataset = pd.merge(gTransactions, gDifficulty, on=['Year', 'Miner Name'], how='outer')
print("gDifficulty")
print(newDataset)
```

```python
#Eliminar Columna del Dataframe
newDataset= newDataset.drop(columns=['Miner Name'])
```

##Numpy

```python
#DataFrame to Numpy
arrayData = newDataset.to_numpy(dtype='int')
print(arrayData)

print("arrayData.shape : " ,arrayData.shape)
```

```python
arrayData.max(axis=1) #0 columnas , 1 filas
# arrayData.max(axis=0)
# arrayData.mean(axis=0)
# arrayData.min(axis=0)
# np.max(arrayData,axis=0)
#arrayData[0]
#np.sum(arrayData[0])
```

```python
df=dfData    
title="Hashrate"
minHrT=3

listBestMiners=df.groupby(['Miner Name'])['Miner Name'].count().to_frame().rename(columns={'Miner Name':'Cantidad'}).reset_index().sort_values(by=['Cantidad'],ascending=False)
s=listBestMiners['Cantidad'].sum()
listBestMiners['Hashrate']=(listBestMiners['Cantidad']/s)*100
listBestMiners= listBestMiners[listBestMiners['Hashrate']>=minHrT]
r=listBestMiners[['Miner Name','Hashrate']]

pHashratecovered= r['Hashrate'].sum()
    
##-------------------    
sns.set(style="whitegrid",font='Times New Roman')
plt.figure( figsize=(10,7.5))
plt.suptitle(title , fontsize=16)    
graph = sns.barplot(x='Hashrate', y="Miner Name", data = r ,  palette='Spectral')

r
```
